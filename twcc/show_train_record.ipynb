```python
import torch
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
checkpoint = torch.load(os.path.join("saves", "checkpoint.pth.tar"))
steps = checkpoint["step"]
best_score = checkpoint["best_score"]
loss_list = checkpoint["loss"]
dice_list = checkpoint["dice"]
savestep = steps / len(loss_list)
```

```python
print(f"Steps: {steps}")
print(f"Checkpoint saved per {savestep:4.0f} steps")
print(f"Best Dice score: {best_score:2.3f}")
```

```python
x = np.arange(1, checkpoint["step"]+1, savestep)
plt.figure(figsize=(20, 5))
plt.plot(x, loss_list, x , dice_list)
plt.ylim(0, 1)
plt.yticks(np.arange(1.1, step=0.1))
plt.legend(("loss", "dice"))
plt.grid()
plt.show()
```
